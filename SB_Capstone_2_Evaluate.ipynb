```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_squared_error
```

```python
df_targets = pd.read_csv('./targets.csv', index_col=0)
df_true = pd.read_csv('./test_targets.csv', index_col=0)
df_predictions = pd.read_csv('./test_predictions.csv', index_col=0)
```

```python
df_predictions.columns = df_targets.columns
df_true.columns = df_targets.columns
```

```python
mean_squared_error(df_true.values, df_predictions.values)
```

```python
mean_squared_error(df_true.iloc[:, 0].values, df_predictions.iloc[:, 0].values)
```

```python
df_mse = pd.DataFrame(index = df_true.columns.values, columns=['class_MSE'])
for c in df_true.columns:
    df_mse.at[c, 'class_MSE'] = mean_squared_error(df_true[c].values, df_predictions[c].values)
```

```python
df_mse['class_MSE'] = df_mse['class_MSE'].astype(float)
```

<img src="./GalaxyClasses.png">

```python
print('The class with the maximum mean square error was ' + 
      '{} with an error of {:0.4f}'.format(df_mse['class_MSE'].idxmax(), df_mse['class_MSE'].max()))
```

```python
print('The class with the minimum mean square error was ' + 
      '{} with an error of {:0.4f}'.format(df_mse['class_MSE'].idxmin(), df_mse['class_MSE'].min()))
```

```python
print('The most frequently represented class in the true data was ' + 
      '{} with an frequency of {:0.4f}'.format(df_true.mean().idxmax(), df_true.mean().max()))
```

```python
print('The least frequently represented class in the true data was ' + 
      '{} with an frequency of {:0.4f}'.format(df_true.mean().idxmin(), df_true.mean().min()))
```

```python
print('The most frequently represented class in the predicted data was ' + 
      '{} with an frequency of {:0.4f}'.format(df_predictions.mean().idxmax(), df_predictions.mean().max()))
```

```python
print('The least frequently represented class in the predicted data was ' + 
      '{} with an frequency of {:0.4f}'.format(df_predictions.mean().idxmin(), df_predictions.mean().min()))
```

```python
df_freq = pd.DataFrame(index=df_true.columns, columns=['avg_freq_true', 'avg_freq_predicted'])
```
